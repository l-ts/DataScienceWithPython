First of all, we import the required modules.

```python
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import statsmodels.graphics as sgr
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.collections import LineCollection
from statsmodels.stats.outliers_influence import summary_table
from statsmodels.formula.api import logit
from pandas.core import datetools

%matplotlib inline
```

We import the csv file.

```python
crd_df = pd.read_csv('creditcard.csv') #read csv file 
crd_df.head(20) #select first 10 records
```

Now we will run logistic regression, because we want to build a model for the binary outcome of fraud detection (0->False, 1->True).

The dependent variable will be the class column.

```python
f = 'Class ~ Time+V1+V2+V3+V4+V5+V6+V7+V8+V9+V10+V11+V12+V13+V14+V15+V16+V17+V18+V19+V20+V21+V22+V23+V24+V25+V26+V27+V28+Amount'
logitfit = smf.logit(formula = str(f), data = crd_df).fit() #fit to our model
```

```python
logitfit.summary() #summary of our model
```

Now we will calculate the probability of each record to be classified as fraud.

```python
y_pred = logitfit.predict() #predict values of fraud (binary) classification
crd_df ['Prediction'] = (y_pred) #store to crd_df dataframe, Prediction column
crd_df.head(10) #select first 10 records
```

```python
x = crd_df[crd_df.Class == 1 ].count()[0] #num of fraud records
y = crd_df[crd_df.Class == 0 ].count()[0] #num of non-fraud records
z = x/(x+y) #proportion of fraud records
print('In our sample the proportion of fraud is:',round(z*100,3),'%.')
```

Since the dataset states that only 0.173% of the sample are frauds, we will get the top 0.173% of our sample and set the prediction to True for these values. For the rest, we will set the prediction of our model to False. Finally we will calculate the accuracy.

Note: Another approach would be to classify according to probability :
crd_df.Prediction >= 0.173% => fraud 
else not fraud.

The method we chose is based on the ranking of the probabilities and not the absolute value, since we know 0.173% is fraud. For example, if our model underestimated the probability, we might ended up getting a lower percentage of frauds than the actual 0.173% we know that is the mean of our population. The same would apply if our model overestimated  the probability value.

```python
limit = crd_df.Prediction.quantile(1-z) #we limit our prediction to top (1-0.00173) quantile
crd_df ['C_Pred'] = np.where(crd_df.Prediction >= limit, True, False) #and set them as fraud predictions
```

We will split into two dataframes, 'positives' and 'negatives' for positive and negative predictions, respectively.

```python
negatives = crd_df[crd_df.C_Pred == False] #negative predictions 
positives = crd_df[crd_df.C_Pred == True] #positive predictions
```

```python
fn = negatives[negatives.Class == 1 ].count()[0] #false negatives
tn = negatives[negatives.Class == 0 ].count()[0] #true negatives
print('False negatives are',fn,'and true negatives are',tn,'.')
```

```python
fp = positives[positives.Class == 0 ].count()[0] #false positives
tp = positives[positives.Class == 1 ].count()[0] #true positives
print('False positives are',fp,'and true positives are',tp,'.')
```

```python
accuracy = (tn+tp)/(fn+tn+fp+tp) #total accuracy
print('Accuracy measured as the ratio of true predictions over total predictions is:',round(accuracy*100,2),'%.')
```

```python
precision = tp/(tp+fp) #precision is percentage of true positives over total positives predicted
print('The precision is',round(precision*100,2),'%.')
```

```python
recall = tp/(tp+fn) #recall is percentage of true positives over all actual positives
print('The recall is',round(recall*100,2),'%.')
```

Now, we would to improve our model so that it is more accurate, and have higher precision and recall. 

One way is to undersample our data. We will split data into two categories according to the amount of the transaction and check whether the fraud rates remains around 0,172% or if it differs.

One hypothesis is that fraud could be much more common for small amounts. Of course the same applies for large amounts. The whole point is that fraud could be detected with different possibilities in distinct subsets of our data.

Since we have no additional information we will try many values for splitting data according to amount quantiles, until we observe a difference in the proportion of fraud ( if we finally do). We decide to check at the tails to see any difference, because usually in statistics they are areas of great interest and may hide extra information.

```python
test_limit = [0.005,0.01,0.05,0.10,0.25,0.50,0.75,0.90,0.95,0.99,0.995,0.999]

for i in test_limit:
    lim = crd_df.Amount.quantile(i)
    cm_crd_df = crd_df[crd_df.Amount > lim ].copy() #copy more than limit
    x_1 = cm_crd_df[cm_crd_df.Class == 0].count()[0] 
    y_1 = cm_crd_df[cm_crd_df.Class == 1].count()[0]
    z_1 = y_1/(x_1+y_1)
    cl_crd_df = crd_df[crd_df.Amount <= lim ].copy() #copy less than limit   
    x_2 = cl_crd_df[cl_crd_df.Class == 0].count()[0]
    y_2 = cl_crd_df[cl_crd_df.Class == 1].count()[0]
    z_2 = y_2/(x_2+y_2)
    print('Fraud for amounts greater than', round(lim,2), ', the proportion of fraud is',round(z_1*100,2),'%.')
    print('For amounts smaller or equal to',round(lim,2),',the proportion of fraud is',round(z_2*100,2),'%.')
    print('Respective quantile point is',i,'.')
    print()
```

Normally, fraud should be around 0.17%.

We notice that for amounts greater than 203.0 , which corresponds to 0.90 quantile, fraud is found at 0.3% of the data, which is almost double as expected. 

Thus, we will split the data into the 0.95 quantile and perform analysis seperately at the two sets, following the same strategy with the initial, in order to compare the results.

Of course, we could split our data into many more sets, achieving higher accuracy but then we would overfit our sample data and probably would have lower accuracy on the population.

```python
lim_final = crd_df.Amount.quantile(0.9) #get 0.9 quantile value
cm_crd_df = crd_df[crd_df.Amount > lim_final ].copy() #first subset
cl_crd_df = crd_df[crd_df.Amount <= lim_final ].copy() #second subset
```

We perform logistic regression at the two datasets independently.

```python
f_m = 'Class ~ Time+V1+V2+V3+V4+V5+V6+V7+V8+V9+V10+V11+V12+V13+V14+V15+V16+V17+V18+V19+V20+V21+V22+V23+V24+V25+V26+V27+V28+Amount'
logitfit_m = smf.logit(formula = str(f_m), data = cm_crd_df).fit() #fit our model
```

```python
f_l = 'Class ~ Time+V1+V2+V3+V4+V5+V6+V7+V8+V9+V10+V11+V12+V13+V14+V15+V16+V17+V18+V19+V20+V21+V22+V23+V24+V25+V26+V27+V28+Amount'
logitfit_l = smf.logit(formula = str(f_l), data = cl_crd_df).fit() #fit our model
```

```python
logitfit_m.summary() #summay for first model
```

```python
logitfit_l.summary() #summary for second model
```

Now, we will caclulate the possibilies for the records to be classified as frauds.

```python
y_pred_l = logitfit_l.predict() #model predict
cl_crd_df['Prediction'] = (y_pred_l) #store result
cl_crd_df.head(10) #select first 10 records
```

```python
y_pred_m = logitfit_m.predict() #model predict
cm_crd_df['Prediction'] = (y_pred_m) #store result
cm_crd_df.head(10) #select first 10 records
```

Now, we will proceed evaluating our model compared to previous one.

```python
x_l = cl_crd_df[cl_crd_df.Class == 1 ].count()[0] #number of fraud records in first dataframe
y_l = cl_crd_df[cl_crd_df.Class == 0 ].count()[0] #number of non-fraud records in first dataframe
z_l = x_l/(x_l+y_l) #percentage of fraud records in first dataframe

x_m = cm_crd_df[cm_crd_df.Class == 1 ].count()[0] #number of fraud records in second dataframe
y_m = cm_crd_df[cm_crd_df.Class == 0 ].count()[0] #number of non-fraud records in second dataframe
z_m = x_m/(x_m+y_m)#percentage of fraud records in second dataframe

print('In the sample with the lower amounts, the percentage of fraud is:',round(z_l*100,3),'%.')
print('In the sample with the higher amounts, the percentage of fraud is:',round(z_m*100,3),'%.')
```

We apply the same strategy as previously based on quantiles (ranking) for each dataframe.

```python
limit_l = cl_crd_df.Prediction.quantile(1-z_l) 
cl_crd_df ['C_Pred'] = np.where(cl_crd_df.Prediction >= limit_l, True, False)
cl_crd_df.head(10)
```

```python
limit_m = cm_crd_df.Prediction.quantile(1-z_m)
cm_crd_df ['C_Pred'] = np.where(cm_crd_df.Prediction >= limit_m, True, False)
cm_crd_df.head(10)
```

Now, we will split the two datasets into positive and negative predictions, respectively.

```python
negatives_l = cl_crd_df[cl_crd_df.C_Pred == False]
positives_l = cl_crd_df[cl_crd_df.C_Pred == True]
negatives_m = cm_crd_df[cm_crd_df.C_Pred == False]
positives_m = cm_crd_df[cm_crd_df.C_Pred == True]
```

```python
fn_l = negatives_l[negatives_l.Class == 1 ].count()[0]
tn_l = negatives_l[negatives_l.Class == 0 ].count()[0]
print('False negatives in lower amounts dataset are',fn_l,'and true negatives are',tn_l,'.')
```

```python
fn_m = negatives_m[negatives_m.Class == 1 ].count()[0]
tn_m = negatives_m[negatives_m.Class == 0 ].count()[0]
print('False negatives in lower amounts dataset are',fn_m,'and true negatives are',tn_m,'.')
```

```python
fp_l = positives_l[positives_l.Class == 0 ].count()[0]
tp_l = positives_l[positives_l.Class == 1 ].count()[0]
print('False positives in higher amounts dataset are',fp_l,'and true positives are',tp_l,'.')
```

```python
fp_m = positives_m[positives_m.Class == 0 ].count()[0]
tp_m = positives_m[positives_m.Class == 1 ].count()[0]
print('False positives in higher amounts dataset are',fp_m,'and true positives are',tp_m,'.')
```

```python
new_accuracy = (tn_l+tn_m+tp_l+tp_m)/(fn_l+fn_m+tn_l+tn_m+fp_l+fp_m+tp_l+tp_m)
print('New accuracy  is:',round(new_accuracy*100,2),'%.')
```

```python
new_precision = (tp_l+tp_m)/(tp_l+tp_m+fp_l+fp_m)
print('The precision is',round(new_precision*100,2),'%.')
```

```python
new_recall = (tp_l+tp_m)/(tp_l+tp_m+fn_l+fn_m)
print('The recall is',round(new_recall*100,2),'%.')
```

Undersampling our initial sample was a promising strategy, but finally it did not work as expected.
