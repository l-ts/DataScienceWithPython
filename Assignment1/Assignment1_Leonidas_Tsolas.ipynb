# Question 1

```python
import tweepy #imports tweety library which connects to twitter api
from twitter_config import config #imports config from file twitter_config.py which contains the keys and tokens

def get_tweets(username,config, max_tweets = 100): #get_tweets is the required function. Default value 100 set for max_tweets 
    CK = config["consumer_key"] 
    CS = config["consumer_secret"]
    AT = config["access_token"]
    ATS = config["access_token_secret"]
    auth = tweepy.OAuthHandler(CK, CS)
    auth.set_access_token(AT, ATS)
    api = tweepy.API(auth)    
    counter = 0 #initiate counter 
    tweets = api.user_timeline(screen_name = username, count = max_tweets)#get the set of all tweets with count set in defintion of get_tweets
    for tweet in tweets: #loop for all tweets fetched in previous line
            if ("RT @") in tweet.text: #"RT @" has been chosen as the expression to search in the text of each tweet
                counter = counter + 1  #if the if statement return true then counter is increased by one            
    return(round(counter/max_tweets,1)) #get_tweets function return proportion of tweets that are replies rounded to first decimal digit
```

```python
import requests
from bs4 import BeautifulSoup

books_in_page = 25 
# total num of books per page, which is default value in Gutenberg's search engine

start = 1 
# inspecting element shows that counting books starts from 1. 

books = [] 
# declaring list of books(initially empty)

try: #if request suceeds do
    while books_in_page < 251:
# there is no information about the total number of results in the first page displayed from search.
# browsing throught the serach result, we notice that there are 243 results.
# Since results increasing step is 25, the comparison with 251 and above will return no results.
# It is easy to notice that if any other number bigger than 251 would be selected, the result would be the same. 

        r = requests.get(f"https://www.gutenberg.org/ebooks/search/?start_index={books_in_page}&query=Charles+Dickens")
        html = r.content
        soup = BeautifulSoup(html, 'html.parser')
        results = soup.find_all("span", {"class":{"title"}}) #find results by class title
        for result in results:
            books.append(result.text)
        books_in_page += 25
        # books in page are increased by 25 in each loop to request next page (passed as a parameter in request URL)
    for book in books:
        print(book)
        # print all the book titles
except requests.exceptions.RequestException as rex:
# in case request is not successful     
    print(rex)
```
